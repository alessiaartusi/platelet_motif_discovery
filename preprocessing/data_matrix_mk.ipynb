```python
import csv
```

```python
d = {}
with open("GSE113182_series_matrix.txt") as f:
    for line in f:
       line = line.split("\t")
       if len(line) > 2:
          k = line[0].replace("!", "")
          v = line[1:]
          v[-1] = v[-1].replace("\n", "")
          d[k] = v

with open('GSE113182_series_matrix.csv', 'w') as f:
    w = csv.writer(f)
    key_list = list(d.keys())
    limit = len(key_list)

    w.writerow(key_list)

    for i in range(20):
      w.writerow(d[x][i] for x in key_list)
```
